## 分析过程
* 获取不到的只能够通过抓包+分析获得，通常是由请求的第一个或第二个文件。
* 翻页以后会有规律的参数，具体如下分析过程

```python
import urllib.request
import random
import re

# 规律总结：发现变化参数
cid = 6636227627315213305
endid = 1582815116256

# 安全一点
uapools = UAPOOL = [
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:60.0) Gecko/20100101 Firefox/60.0",
    "Mozilla/5.0 (Windows NT 6.1) AppleWebKit/536.5",
    "Mozilla/5.0 (Windows NT 6.1; WOW64; rv:48.0) Gecko/20100101 Firefox/48.0",
    "Mozilla/4.0 (compatible; MISE 7.0; Windows NT 5.1; Maxthon 2.0)"
]

def UA():
    opener = urllib.request.build_opener()
    thisua = random.choice(uapools)
    ua = ("User-Agent",thisua)
    opener.addheaders = [ua]
    urllib.request.install_opener(opener)
    
def save(filename, contents):
      fh = open(filename, 'w', encoding='utf-8')
      fh.write(contents)
      fh.close()

for i in range(0,100):
    print("第"+str(i+1)+"页")
    url = "https://video.coral.qq.com/varticle/4825888381/comment/v2?callback=_varticle4825888381commentv2&orinum=10&oriorder=o&pageflag=1&cursor="+str(cid)+"&scorecursor=0&orirepnum=2&reporder=o&reppageflag=1&source=132&_="+str(endid+i)
    UA()
    data = urllib.request.urlopen(url).read().decode("utf-8","ignore")
    pat = '"content":"(.*?)"'
    comments = re.compile(pat,re.S).findall(data)
    for item in comments:
        string = str(eval('u"'+str(item)+'"'))
        newdata =  string +'\n----\n'
        print(string)
        print('----')
        save('D:\\三生三世评论.txt', newdata)
    print('ok'+str(i+1))
    pat2 = '"last":"(.*?)"'
    cid = re.compile(pat2,re.S).findall(data)[0]
```
