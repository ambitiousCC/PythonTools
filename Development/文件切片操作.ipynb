## 背景
相对文件进行切片操作

```python
f = open('hamlet.txt','rb')
```

```python
for line in f:
    print(line)
f.seek(0)
```

```python
from itertools import islice
```

```python
islice(f,100,300)
```

切片操作

```python
for line in islice(f,100,300):
    print(line)
```

会消耗迭代器原来的对象

```python
l = range(20)
l
```

```python
t = iter(l)
```

```python
for x in islice(t,5,10):
    print(x)
```

```python
for x in t:
    print(x)
```
