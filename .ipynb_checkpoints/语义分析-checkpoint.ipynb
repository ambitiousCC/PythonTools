````python
def dh_list(file):
    with open(file,"r+",encoding="utf-8") as f: 
        result=[]
        for line in f.readlines():#readlines以列表输出文件内容
            line=line.replace("\t"," d1(); ")#改变元素，去掉，和换行符\n,tab键则把逗号换成"/t",空格换成" "
            result.append(line)
    return result

print(dh_list('光遇乐谱.txt'))
#结果：['我是哪吒','hello world','nice to meet you']
````

````python
import os

resultList=[]
for i in range(1,2):
    with open("500.txt","r+",encoding="utf-8") as f: 
        result=[]
        for line in f.readlines():#readlines以列表输出文件内容
            line=line.replace("\t"," deplay1(); ")#改变元素，去掉，和换行符\n,tab键则把逗号换成"/t",空格换成" "
            result.append(line)

    #with open((os.path.join('光遇乐谱.txt')), 'r') as f:
    #	data=f.readlines()
    for line in result:
        odom=line.split()
        num=map(str,odom)
        lst=(list(num))
    Len=len(lst)
     
    for i in range(0,Len):
        if lst[i]=='1':
            lst[i]="a1();"
        elif lst[i]=='2':
            lst[i]="a2();"
        elif lst[i]=='3':
            lst[i]="a3();"
        elif lst[i]=='4':
            lst[i]="a4();"
        elif lst[i]=='5':
            lst[i]="a5();"
        elif lst[i]=='6':
            lst[i]="a6();"
        elif lst[i]=='7':
            lst[i]="a7();"
        elif lst[i]=='11':
            lst[i]="b1();"
        elif lst[i]=='22':
            lst[i]="b2();"
        elif lst[i]=='33':
            lst[i]="b3();"
        elif lst[i]=='44':
            lst[i]="b4();"
        elif lst[i]=='55':
            lst[i]="b5();"
        elif lst[i]=='66':
            lst[i]="b6();"
        elif lst[i]=='77':
            lst[i]="b7();"
        elif lst[i]=='111':
            lst[i]="c1();"
    print("list.len=",Len)
    resultList.append(lst)
print(resultList[0])
````
