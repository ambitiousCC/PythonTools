## DataFrame 基本操作

## 转换时间字符串为标准时间 4-41

```python
import pandas as pd
order = pd.read_table('./exe/meal_order_info.csv',
                     sep=',',encoding='gbk')
print('转换前的lock_time类型：',order['lock_time'].dtypes)
order['lock_time'] = pd.to_datetime(order['lock_time'])
print('转换以后:',order['lock_time'].dtypes)
```

```python
print('最小时间：',pd.Timestamp.min)
```

```python
print('最大时间：',pd.Timestamp.max)
```

## 字符串转化对应数据关系索引值

```python
dateIndex = pd.DatetimeIndex(order['lock_time'])
print('转换DatetimeIndex之后数据的类型',type(dateIndex))
```

```python
periodIndex = pd.PeriodIndex(order['lock_time'],freq='S')
print("转换为PeriodIndex之后的数据的类型",type(periodIndex))
```

## 提取datetime数据中的时间序列数据

```python
year1 = [i.year for i in order['lock_time']]
print('lock_time中的年份数据前5个位：',year1[:5])
```

```python
month1 = [i.month for i in order['lock_time']]
print('lock_time中的月份数据前五个位：',month1[:5])
```

```python
weekday1 = [i.weekday_name for i in order['lock_time']]
print('lock_time中星期名称数据前5个为：',weekday1[:5])
```

```python
print('dataIndex中星期名称数据前5个位：',dateIndex.weekday_name[:5])
print('periodIndex中的星期标号数据的前5个数据为：',periodIndex.weekday[:5])
```

## 加减时间

```python
time1 = order['lock_time'] + pd.Timedelta(days = 1)
print('lock_time加上一天前前5数据：',order['lock_time'][:5])
print('lock_time加上一天前前5行数据为：',time1[:5])
```

```python
timeDelta = order['lock_time'] - pd.to_datetime('2017-1-1')
print(timeDelta[:5])
print('lock_time减去2017年1月1日0.0.0:',timeDelta.dtypes)
```
